```python
def get_nth_key(dict, n):
    for i, key in enumerate(dict.keys()):
        if i == n:
            return key

import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("../data/nsaf_data_new_timecleaned.csv")

times=data["Cleaned Time Period"].to_list()

times_int=[]
for time in times:
    try:
        time_int=int(time)
        times_int.append(time_int)
    except:
        pass
```

```python
dict={"1st":0, "2nd":0, "3rd":0, "4th":0, "5th":0, "6th":0, "7th":0, 
      "8th":0, "9th":0, "10th":0, "11th":0, "12th":0, "13th":0, "14th":0,"15th":0,
      "16th":0,"17th":0,"18th":0,"19th":0,"20th":0,"21st":0}

for time in times_int:
    time_century = time//100
    key = get_nth_key(dict, time_century)
    dict[key] = dict.get(key) + 1

x1=list(dict.keys())
y1=list(dict.values())

fig, ax = plt.subplots()
ax.plot(x1,y1)
plt.setp(ax.get_xticklabels()[1::2], visible=False)
ax.set(xlabel='Century', ylabel='Number of Art Pieces',
       title='Stolen Art by Time of Creation')
plt.show()
```

```python
data = pd.read_csv("../data/nsaf_data_known_dates.csv")
times=data["Cleaned Time Period"].to_list()

times_int=[]
for time in times:
    try:
        time_int=int(time)
        times_int.append(time_int)
    except:
        pass

dict2={"1900s":0, "1910s":0, "1920s":0, "1930s":0, "1940s":0, "1950s":0, "1960s":0, 
      "1970s":0, "1980s":0, "1990s":0, "2000s":0, "2010s":0, "2020s":0}

after_1900 = []

for time in times_int:
    if time>=1900:
        after_1900.append(time)

for t in after_1900:
    new_time=t-1900
    index=new_time//10
    key = get_nth_key(dict2, index)
    dict2[key] = dict2.get(key) + 1

x2=list(dict2.keys())
y2=list(dict2.values())

fig, main_ax = plt.subplots()
main_ax.plot(x2,y2)
plt.setp(main_ax.get_xticklabels()[1::2], visible=False)
main_ax.set(xlabel='Decade', ylabel='Number of Art Pieces',
       title='Stolen Art by Time of Creation 1900-Present')
plt.show()
```
