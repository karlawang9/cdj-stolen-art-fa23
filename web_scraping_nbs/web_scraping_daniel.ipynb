```python
import requests
from bs4 import BeautifulSoup
import re

import pandas as pd
import numpy as np
```

```python
html_source = ""

# ORIGINAL CODE
# with open("p1.html", "r") as reader:
#     html_source = reader.read()

with open("../fbi_nsaf_pages/p1.html", "r") as reader:
    html_source = reader.read()
```

```python
data = {
    "Title" : [],
    "Category" : [], 
    "Reference Number" : [], 
    "Maker/Artist" : [], 
    "Materials" : [], 
    "Measurements" : [], 
    "Time Period" : [],
    "Additional Information" : [],
    "Image Link":[ ]
}

soup = BeautifulSoup(html_source, "html.parser")


grid_items = soup.find_all("li", class_="grid-item")

for item in grid_items:
  img_tag = item.find("img")
  if img_tag:
    img_src = img_tag.get("src")
  else:
    img_src = "NaN"

  title_tag = item.find("h3")
  if title_tag:
    title = title_tag.text
  else:
    title = "NaN"

  category_element = item.find('span', text='Category')
  if category_element:
    category_element = category_element.find_next('span')
    category = category_element.get_text()
  else:
    category = "NaN"

  ref_element = item.find('span', text='Reference Number')
  if ref_element:
    ref_element = ref_element.find_next('span')
    reference = ref_element.get_text()
  else:
    reference = "NaN"

  maker_element = item.find('span', text='Maker/Artist')
  if maker_element:
    maker_element = maker_element.find_next('span')
    maker = maker_element.get_text()
  else:
    maker = "NaN"

  materials_element = item.find('span', text='Materials')
  if materials_element:
    materials_element = materials_element.find_next('span')
    materials = materials_element.get_text()
  else:
    materials = "NaN"

  measurements_element = item.find('span', text='Measurements')
  if measurements_element:
    measurements_element = measurements_element.find_next('span')
    measurements = measurements_element.get_text()
  else:
    measurements = "NaN"

  time_element = item.find('span', text='Time Period')
  if time_element:
    time_element = time_element.find_next('span')
    time = time_element.get_text()
  else:
    time = "NaN"

  additional_element = item.find('span', text='Additional Information')
  if additional_element:
    additional_element = additional_element.find_next('span')
    additional = additional_element.get_text()
  else:
    additional = "NaN"

  data['Title'] += [title]      
  data['Category'] += [category]
  data['Reference Number'] += [reference]
  data['Maker/Artist'] += [maker]
  data['Materials'] += [materials]
  data['Measurements'] += [measurements]
  data['Time Period'] += [time]
  data['Additional Information'] += [additional]
  data['Image Link'] += [img_src]

df = pd.DataFrame(data)

df.head(20)
```
