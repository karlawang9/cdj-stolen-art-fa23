```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# ORIGINAL CODE
# with open("p1.html", "r") as reader:
#   html_source = reader.read()

with open("../fbi_nsaf_pages/p1.html", "r") as reader:
  html_source = reader.read()

page = BeautifulSoup(html_source, "html.parser")
```

```python
image = []
name = []
category = []
referenceNumber= []
makerArtist = []
materials = []
measurements = []
timePeriod = []
additionalInfo = []
```

```python
dictionary = {"Image": image, "Name": name, "Category": category, "Reference Number": referenceNumber, "Maker/Artist": makerArtist, "Materials": materials, "Measurements": measurements, "Time Period": timePeriod, "Additional Information": additionalInfo}
resultCategories = [category, referenceNumber, makerArtist, materials, measurements, timePeriod, additionalInfo]
namesCategories = ["Category", "Reference Number", "Maker/Artist", "Materials", "Measurements", "Time Period", "Additional Information"]
```

```python
items = page.find_all("li", {"class":"grid-item"})
for item in items:
    #image
    img = item.find("img")
    try:
        image.append(img['src'])
    except:
        image.append("")
        
    #name
    n = item.find("h3")
    try:
        name.append(n.get_text())
    except:
        name.append("")
    
    #ul
    list = item.find("ul")
    listElements = list.find_all("li")
    for element in listElements:
        span = element.find("span")
        for na in namesCategories:
            if (span.get_text()==na):
                next = span.find_next("span")
                text = next.get_text()
                dictionary.get(na).append(text)
                break
    
    length=max(len(image), len(name))
    
    for i in resultCategories:
        if (len(i)!=length):
            i.append("")
```

```python
df = pd.DataFrame(dictionary)
df
```
